```python
import fast_voxel_traversal as fvt
```

```python
import numpy as np

bounds = ((-100.0, 0.0, 0.0), (100.0, 200.0, 300.0))
shape = (50, 40, 30)
grid = fvt.Grid(bounds=bounds, shape=shape)

origin = (-10.0, -20.0, 2.5)
direction = (1.0, 2.0, 0.5)
ray = fvt.Ray(origin=origin, direction=direction)

ray_voxels = ray.time_grid(grid)
min = np.nanmin(ray_voxels)
ray_voxels = ray_voxels + 1 - min
ray_voxels = np.nan_to_num(ray_voxels)
ray_voxels = np.interp(ray_voxels, (0, np.max(ray_voxels)), (0, 1))
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
# %matplotlib qt

fig, ax = plt.subplots(subplot_kw={'projection': '3d'})
grid.plot(ray_voxels, ax=ax, show=False)
plt.show()
```
